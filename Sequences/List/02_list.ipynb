  <p style="margin:0 0 12px 0;">

    <li>In a <strong style="color:#f38ba8">deque</strong>, <code>my_deque[5000]</code> forces Python to start at one end and jump from block to block until it counts up to 5,000 — <strong style="color:#f38ba8">O(n)</strong></li>
  </ul>

    <strong style="color:#cba6f7">Rule of thumb:</strong> the <code>*</code> multiplier is perfectly fine for <strong style="color:#a6e3a1">flat, immutable data</strong> (like <code>[0] * 5</code> or <code>["abc"] * 10</code>). But the moment you have <strong style="color:#f38ba8">nested mutable objects</strong> (like lists or dictionaries), always reach for a <strong style="color:#89b4fa">comprehension</strong> to ensure everything gets its own unique space in memory.
  </p>

```python
# List while iterating over it
# Imagine you want to remove all even numbers from a list: [10, 20, 30, 40].
numbers = [10, 20, 30, 40]
for num in numbers:
    if num%2==0:
        numbers.remove(num)
print(numbers)
# Expected: []
# Actual Output: [20, 40]  <-- Wait, what?!
```

<div style="background:#1e1e2e; padding:16px 20px; border-radius:8px; font-family:monospace; color:#cdd6f4; line-height:1.8">
  <p style="margin:0 0 12px 0;">Here is exactly what Python did step-by-step:</p>
  <ol style="margin:0 0 14px 0; padding-left:22px; line-height:2.4">
    <li><strong style="color:#89b4fa">Iteration 0</strong> — pointer at index <code>0</code>, item is <code>10</code>. It's even, Python removes it. List shifts left → <code>[20, 30, 40]</code></li>
    <li><strong style="color:#89b4fa">Iteration 1</strong> — pointer increments to index <code>1</code>. In <code>[20, 30, 40]</code>, index <code>1</code> is now <code>30</code> — <strong style="color:#f38ba8"><code>20</code> was completely skipped</strong> because it shifted into index <code>0</code>. Python sees <code>30</code>, removes it → <code>[20, 40]</code></li>
    <li><strong style="color:#89b4fa">Iteration 2</strong> — pointer increments to index <code>2</code>. List is <code>[20, 40]</code> — no index <code>2</code> exists, loop gracefully exits</li>
  </ol>
  <p style="margin:0; border-left:4px solid #f38ba8; padding-left:12px">
    By shifting elements left while moving the pointer right, you end up <strong style="color:#f38ba8">jumping over every other item</strong>.
  </p>
</div>

```python
# Fix 1: Iterating Over a Copy (for x in lst[:])
# When you write lst[:] (or lst.copy()), 
# Python duplicates the shallow structure of the list in memory.
numbers = [10, 20, 30, 40]

# We pull numbers from the copy, but mutate the original!
for num in numbers[:]:
    if num%2 == 0:
        numbers.remove(num)
print(numbers)  # Correct
```

<div style="background:#1e1e2e; padding:16px 20px; border-radius:8px; font-family:monospace; color:#cdd6f4; line-height:1.8">
  <p style="margin:0;">
    The <code>for</code> loop tracks its pointer along the <strong style="color:#89b4fa">copy</strong>, which never changes size or shape. Even though you are deleting elements from the original list, the loop's pointer steadily marches through index <code>0</code>, <code>1</code>, <code>2</code>, and <code>3</code> of the copy — <strong style="color:#a6e3a1">without ever skipping a beat</strong>.
  </p>
</div>

```python
# Fix 2: List Comprehension
```

<div style="background:#1e1e2e; padding:16px 20px; border-radius:8px; font-family:monospace; color:#cdd6f4; line-height:1.8">
  <p style="margin:0 0 12px 0; color:#a6e3a1; font-weight:bold;">Fix 2 — List Comprehension (Best Practice)</p>
  <p style="margin:0 0 12px 0;">
    Slicing a copy works, but still suffers from poor performance — every <code>numbers.remove(num)</code> call forces Python to search the list and shift elements, making it <strong style="color:#f38ba8">O(n²)</strong>.
  </p>
  <p style="margin:0; border-left:4px solid #a6e3a1; padding-left:12px">
    The most efficient and elegant solution is to turn the logic upside down: instead of thinking about <strong style="color:#f38ba8">what to remove</strong>, use a comprehension to specify <strong style="color:#a6e3a1">what to keep</strong>.
  </p>
</div>

```python
numbers = [10, 20, 30, 40]
# Build a brand new list containing only what you want
numbers = [num for num in numbers if num%2!=0]
print(numbers)
```

<div style="background:#1e1e2e; padding:16px 20px; border-radius:8px; font-family:monospace; color:#cdd6f4; line-height:1.8">
  <p style="margin:0 0 12px 0;">
    This avoids the mutation trap entirely — it doesn't alter the list while iterating. It reads smoothly from the old list, pipes the valid data into a brand new list, and does it in a highly optimized <strong style="color:#a6e3a1">O(n) single pass</strong>.
  </p>
  <p style="margin:0; border-left:4px solid #cba6f7; padding-left:12px">
    <strong style="color:#cba6f7">Rule of Thumb:</strong> if you are filtering a list, <strong style="color:#a6e3a1">always use a list comprehension</strong>. Only reach for the slice-copy trick (<code>lst[:]</code>) if you absolutely must mutate a massive list in-place and cannot reassign the variable name.
  </p>
</div>

<div style="background:#1e1e2e; padding:16px 20px; border-radius:8px; font-family:monospace; color:#cdd6f4; line-height:1.8">

```python
# 2. Using .extend() with a String (The Trap)
# If you pass that same string to .extend(), 
# Python iterates through the string character-by-character.
cities = ["Tokyo", "Paris"]
cities.extend("Delhi")
cities
```

<div style="background:#1e1e2e; padding:16px 20px; border-radius:8px; font-family:monospace; color:#cdd6f4; line-height:1.8">

  <p style="margin:0 0 16px 0; color:#cba6f7; font-weight:bold;">The Math — O(n) vs O(n²)</p>
  <p style="margin:0 0 14px 0;">What actually happens under the hood when you scale a growth loop to 10,000 iterations:</p>

  <p style="margin:0 0 8px 0; color:#f38ba8; font-weight:bold;">1. The Slow Way — <code>result = result + [x]</code></p>
  <p style="margin:0 0 8px 0;">On the first iteration, Python copies 1 item. On the second, 2 items. On the 10,000th iteration, it copies 9,999 items just to add one new one. The total operations:</p>
  <div style="background:#181825; border-radius:6px; padding:12px 16px; margin:0 0 8px 0; border-left:4px solid #f38ba8; text-align:center;">
    <span style="color:#cdd6f4;">1 + 2 + 3 + ... + n  =  n(n + 1) / 2  ≈  n² / 2</span>
  </div>
  <p style="margin:0 0 16px 0;">Growing a list to 10,000 items using <code>+</code> requires roughly <strong style="color:#f38ba8">50 million copy operations — O(n²)</strong>.</p>

  <p style="margin:0 0 8px 0; color:#a6e3a1; font-weight:bold;">2. The Fast Way — <code>result.append(x)</code> or <code>result.extend([x])</code></p>
  <p style="margin:0 0 8px 0;">Python lists are <strong style="color:#89b4fa">dynamic arrays</strong> — they allocate more memory than they currently need (called <strong style="color:#89b4fa">over-allocation</strong>). When you use <code>.append()</code> or <code>.extend()</code>, Python usually just drops the new item into pre-allocated empty space. It only resizes and copies occasionally when that extra space completely runs out.</p>
  <p style="margin:0; border-left:4px solid #a6e3a1; padding-left:12px">
    Amortized over the entire loop, each addition takes <strong style="color:#a6e3a1">O(1) constant time</strong>. Growing the list to 10,000 items takes roughly <strong style="color:#a6e3a1">10,000 operations — O(n)</strong>.
  </p>

</div>

```python
list_a = [1,2]
list_b = list_a

# Using += mutates the list in place
list_a += [3, 4]

print(list_a)
print(list_b)
# (list_b saw the changes because it's the same object!)
```

```python
list_a = [1,2]
list_b = list_a

# Using + creates a brand NEW list and reassigns list_a
list_a = list_a + [3,4]
print(list_a)
print(list_b)
```

5 (Hard) — Running maximum. Write running_max(nums) returning a list where each position holds the max seen so far. running_max([3,1,4,1,5,9,2,6]) → [3,3,4,4,5,9,9,9].
Hint: accumulator pattern. Track the current max, update it each step (cur = max(cur, n)), append cur each iteration. What's a safe initial value? (Think float('-inf'), or seed from the first element.)

```python
def running_max(nums):
    cur = []
    for n in nums:
        try:
            cur = max(cur, n)
            cur.append(n)
        except Exception as e:
            print(e)
    print(cur)
nums = [3,1,4,1,5,9,2,6]
running_max(nums)
```

```python
def running_max(nums):
    out = []
    cur = float('-inf')     # the result list
    for n in nums:          # the running max - start lower than any number
        cur = max(cur, n)   # update: is this n bigger than what we've seen?
        out.append(cur)     # record the current max at this position
    return out
    
```

```python
print(running_max([3,1,4,4,1,5,9,2,6]))
```

```python
# C2 (Easy/Medium). Two-sum: return indices of two numbers adding to target. ([2,7,11,15], 9) → [0,1].
def two_sum(nums, target):
    seen = {}
    for i,n in enumerate(nums):
        if target - n in seen:
            return [seen[target - n], i]
        seen[n] = i
    return None
```

```python
print(two_sum([2,7,11,15], 9))
```

```python
# C3 (Medium). Move all zeros to the end, preserving the order of non-zeros. [0,1,0,3,12] → [1,3,12,0,0].
def move_zeros(nums):
    nonzero = [x for x in nums if x != 0]
    return nonzero + [0]*(len(nums) - len(nonzero))
nums = [0,1,0,3,12]    
move_zeros(nums)
```

<div style="background:#1e1e2e; padding:16px 20px; border-radius:8px; font-family:monospace; color:#cdd6f4; line-height:1.8">
  <p style="margin:0 0 12px 0;">
    Filter non-zeros (comprehension), then pad with the right count of zeros — <strong style="color:#a6e3a1">O(n)</strong>.
  </p>
  <p style="margin:0; border-left:4px solid #cba6f7; padding-left:12px">
    An in-place <strong style="color:#89b4fa">two-pointer version</strong> exists if asked to avoid extra space.
  </p>
</div>

```python
# C4 (Medium). Chunk a list into sublists of size n. ([1..7], 3) → [[1,2,3],[4,5,6],[7]].
def chunk(lst, n):
    return [lst[i:i+n] for i in range(0, len(lst), n)]
chunk([1,2,3,4,5,6,7], 3)
```

```python
# C5 (Medium). Rotate a list right by k. ([1,2,3,4,5], 2) → [4,5,1,2,3].
def rotate(nums, k):
    k %= len(nums)
    return nums[-k:] + nums[:-k]
rotate([1,2,3,4,5],2)
```

  <p style="margin:0 0 12px 0;">
    Slicing makes this elegant: take the last <code>k</code>, put the rest after. <strong style="color:#a6e3a1">O(n)</strong>.
  </p>
  <p style="margin:0; border-left:4px solid #cba6f7; padding-left:12px">
    <code>k %= len</code> handles rotations larger than the list.

```python
# C6 (Hard). Flatten an arbitrarily nested list. [1,[2,[3,[4,5]]],6] → [1,2,3,4,5,6].
def flatten(lst):
    out = []
    for x in lst:
        if isinstance(x, list):
            out.extend(flatten(x))
        else:
            out.append(x)
    return out
flatten([1,[2,[3,[4,5]]],6])
```

<div style="background:#1e1e2e; padding:16px 20px; border-radius:8px; font-family:monospace; color:#cdd6f4; line-height:1.8">
  <p style="margin:0 0 12px 0;">
    Recursion handles unknown depth — recurse on lists (extend with the flattened result), append non-lists.
  </p>
  <p style="margin:0; border-left:4px solid #cba6f7; padding-left:12px">
    The single-level comprehension <code>[x for sub in nested for x in sub]</code> only handles one level — this is the <strong style="color:#cba6f7">general case</strong>.
  </p>
</div>

```python
# C7 (Hard). Maximum subarray sum (contiguous). [-2,1,-3,4,-1,2,1,-5,4] → 6.
def max_subarray(nums):
    best = cur = nums[0]
    for n in nums[1:]:
        cur = max(n, cur+n)         # extend current run, or start fresh at n
        best = max(best, cur)       # track the best seen
    return best

max_subarray([-2,1,-3,4,-1,2,1,-5,4])
```

<div style="background:#1e1e2e; padding:16px 20px; border-radius:8px; font-family:monospace; color:#cdd6f4; line-height:1.8">
  <p style="margin:0 0 12px 0;">
    <strong style="color:#cba6f7">Kadane's algorithm</strong> — the two-state accumulator from Exercise Problem 5, applied.
  </p>
  <ul style="margin:0 0 12px 0; padding-left:20px; line-height:2.2">
    <li><code>cur</code> — best sum ending here</li>
    <li><code>best</code> — best overall</li>
  </ul>
  <p style="margin:0; border-left:4px solid #a6e3a1; padding-left:12px">
    <strong style="color:#a6e3a1">O(n)</strong> — and a classic interview question.
  </p>
</div>

```python
# C8 (Hard). Group consecutive runs of integers. [1,2,3,5,6,8] → [[1,2,3],[5,6],[8]].
def group_consecutive(nums):
    if not nums: return []
    groups = [[nums[0]]]
    for n in nums[1:]:
        if n == groups[-1][-1] + 1:
            groups[-1].append(n)
        else:
            groups.append([n])
    return groups
group_consecutive([1,2,3,5,6,8])
```

    Track the last element of the current group (<code>groups[-1][-1]</code>) — if the next number is consecutive, extend that group; else open a new one. <strong style="color:#a6e3a1">O(n)</strong>.
