```python
# 19 — two-pointer in-place reverse (the canonical two-pointer move)
arr = [1,2,3,4,5]
i,j = 0, len(arr)-1
while i<j:
    arr[i], arr[j] = arr[j], arr[i]
    i += 1; j -= 1
print(arr)
```

```python
# 3 — sort / sorted / key / multi-key, one progression
raw = [88, 92, 75, 100]
new_sorted = sorted(raw)            # NEW list; raw untouched
# new_sorted -> [75,88,92,100],     raw still [88,92,75,100]

raw.sort(reverse=True)              # in place, descending
# raw -> [100, 92, 88, 75]

records = [("Alice",88),("Bob",92),("Cara",88)]
sorted(records, key=lambda r: r[1]) # by score
# -> [('Alice',88),('Cara',88),('Bob',92)]    Alice before Cara = STABLE

sorted(records, key=lambda r: (-r[1], r[0]))    # score desc, then name asc
# -> [('Bob',92),('Alice',88),('Cara',88)]
```

```python
# 4 — enumerate + zip together
# The idiomatic way to walk two parallel lists with a rank counter:
students = ["Alice", "Bob", "Cara"]
grades = [88, 92, 75]
for rank, (name, grade) in enumerate(zip(students, grades), start=1):
    print(f"  #{rank}: {name} = {grade}")
```

    <li><strong style="color:#89b4fa"><code>*</code> repeat</strong> — flat values only</li>
  </ul>
  <p style="margin:0; border-left:4px solid #f38ba8; padding-left:12px">
    <strong style="color:#f38ba8">Never use <code>*</code> for nested</strong> — aliasing trap.
  </p>
</div>

```python
# 6 — End-to-end record processing
# Everything from both theory chunks in one realistic flow: 
# filter, sort, enumerate, aggregate.
data = [("err",5), ("ok",12), ("err",3), ("warn",8), ("ok",1)]
filtered = [rec for rec in data if rec[0] != "ok"]      # drop "ok" events
filtered.sort(key=lambda r: r[1], reverse=True)         # by count, descending
filtered
```

```python
for i, (level, count) in enumerate(filtered, 1):
    print(f"  {i}. {level:5s} -> {count}")
total = sum(count for _, count in data)                 # aggregate over all
```

```python
# 5 — Sorting mixed types raises TypeError (Python 3)
try:
    [3, "a", 1].sort()
except Exception as e:
    print(e)
```

<div style="background:#1e1e2e; padding:16px 20px; border-radius:8px; font-family:monospace; color:#cdd6f4; line-height:1.8">
  <p style="margin:0 0 12px 0;">
    Python 3 won't compare across incompatible types — a list mixing ints and strings can't be sorted directly.
  </p>
  <p style="margin:0 0 12px 0; border-left:4px solid #f38ba8; padding-left:12px">
    <strong style="color:#f38ba8">Python 2</strong> would sort it with arbitrary results — a notorious source of silent bugs, now correctly an error in Python 3.
  </p>
  <p style="margin:0; border-left:4px solid #a6e3a1; padding-left:12px">
    <strong style="color:#a6e3a1">Fix:</strong> give a <code>key</code> that maps everything to one comparable type — e.g. <code>key=str</code>.
  </p>
</div>

```python
mixed_list = [3, "a", 1]
mixed_list.sort(key=str)
mixed_list
```

<div style="background:#1e1e2e; padding:16px 20px; border-radius:8px; font-family:monospace; color:#cdd6f4; line-height:1.8">
  <p style="margin:0 0 12px 0;">
    The <code>key</code> parameter tells the sorting algorithm: <em>"don't look at the actual items — run each item through this function first, and sort based on the results."</em>
  </p>
  <p style="margin:0 0 12px 0;">
    The <code>key</code> argument expects a <strong style="color:#89b4fa">callable</strong> (a function, a lambda, or a method) that takes one argument and returns one value.
  </p>
  <p style="margin:0 0 12px 0; border-left:4px solid #cba6f7; padding-left:12px">
    <code>key=str</code> — temporarily treats every element as a string during comparison.
  </p>
  <p style="margin:0; border-left:4px solid #a6e3a1; padding-left:12px">
    <strong style="color:#a6e3a1">Note:</strong> the original items keep their original types — integer stays integer, string stays string. Only the <em>comparison</em> used their string representations.
  </p>
</div>

<div style="background:#1e1e2e; padding:16px 20px; border-radius:8px; font-family:monospace; color:#cdd6f4; line-height:1.8">
  <p style="margin:0 0 12px 0;">
    2. Using <code>lambda</code> for custom logic — what if you want all numbers to come first, and all strings to come second?
  </p>
  <p style="margin:0; border-left:4px solid #89b4fa; padding-left:12px">
    Pass a <code>lambda</code> that returns a <strong style="color:#89b4fa">tuple</strong>. Python sorts tuples element by element — so the first value in the tuple acts as the primary sort key, and the second as the tiebreaker.
  </p>
</div>

```python
mixed_list = [3, "b", 1, "a"]

# Tuple sorting: (is_it_a_string, the_value)
# False (0) comes before True (1), so ints go first!
mixed_list.sort(key=lambda x: (isinstance(x, str), x))

print(mixed_list)
```

```python
# Common Scenarios - To use 'Key'
# Scenario A: Sorting Objects or Dictionaries
# If you have complex data, like a list of dictionaries, 
# you can sort by a specific key.
players = [
    {"name": "Alice", "score":88},
    {"name": "Bob", "score": 95},
    {"name": "Charlie", "score": 70}
]

# Sort by score (highest to lowest)
players.sort(key=lambda players: players["score"], reverse=True)
players
```

```python
# Scenario B: Case-Insensitive String Sorting
# By default, Python sorts uppercase letters before lowercase letters 
# ("Zoo" comes before "apple"). You can fix this with str.lower.
words = ["Zoo", "apple", "Banana"]
words.sort(key=str.lower)
words
```

```python
words = ["Zoo", "apple", "Banana"]
sorted_words = sorted(words, key=str.lower)
sorted_words
```

```python
# Scenario C: Sorting by Length
# Want to sort a list of phrases by how long they are?
phrases = ["python programming", "ai", "code"]
sorted_phrases = sorted(phrases, key = lambda phrase: (len(phrase), phrase), reverse=True)
sorted_phrases
```

```python
prhases = ["python programming", "ai", "code"]
sorted_phrases = sorted(phrases, key = len, reverse=True)
sorted_phrases
```

```python
prhases.sort(key=len)
prhases
```

```python
# If you are sorting large amounts of data by dictionary keys or object attributes, 
# using lambda can sometimes be a bit slow. 
# Python has a built-in module called operator designed to make this cleaner and faster.
```

```python
# 8b — append vs extend (the daily confusion)
[1,2].append([3,4])    # [1, 2, [3, 4]]  -- adds the LIST as ONE element
[1,2].extend([3,4])    # [1, 2, 3, 4]    -- adds EACH element
[1,2].append("ab")     # [1, 2, 'ab']    -- string as one element
[1,2].extend("ab")     # [1, 2, 'a', 'b'] -- ITERATES the string!
```
