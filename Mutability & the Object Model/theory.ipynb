<h1 class="main-title">🐍 Session 1 — Mutability &amp; the Object Model</h1>
<div class="info-box"><strong>Part 1:</strong> Theory → Example → Edge Cases</div>

<hr class="divider">

<!-- ══ SECTION 1: THEORY ══ -->
<div class="part-header"><h2>1. Theory</h2></div>

<h3 class="sub">🔹 The One Idea Everything Rests On</h3>

<div class="theory-box">
  <p style="margin:0 0 10px 0">In Python, a variable is not a box that holds a value. It is a <strong style="color:#89b4fa">name tag attached to an object</strong>.</p>
  <p style="margin:0">This is the single mental model that explains every "weird" behavior in this entire session.</p>
</div>

<h3 class="sub">🔹 Power BI / SQL Analogy</h3>
<p style="color:#cdd6f4;font-family:'Segoe UI',sans-serif;font-size:0.93em;line-height:1.7;margin:8px 0">Think of a SQL VIEW. The view name (<code style="background:#313244;padding:2px 6px;border-radius:4px;color:#a6e3a1;font-family:'Courier New',monospace">vw_Sales</code>) doesn't contain data — it points at an underlying table. Two views can point at the same table. Drop a view, the table survives. Change the table, every view sees the change. Python names work exactly like that: the name points at an object living in memory, and multiple names can point at the same object.</p>

<div class="info-box">💡 Another anchor: in Power BI, a measure reference vs. the measure itself. <strong>Total Sales</strong> the name is just a pointer; the actual DAX object lives once and many visuals reference it.</div>

<h3 class="sub">🔹 Three Pieces of Vocabulary</h3>
<p style="color:#cdd6f4;font-family:'Segoe UI',sans-serif;font-size:0.93em;line-height:1.7;margin:8px 0">Every value in Python is an object, and every object has three things:</p>

<div class="vocab-box">
  <div class="vocab-row">
    <span class="vocab-term">Identity</span><span class="vocab-dash">—</span>
    a unique address in memory, never changes during the object's life. You read it with <code style="background:#313244;padding:2px 6px;border-radius:4px;color:#a6e3a1;font-family:'Courier New',monospace">id()</code>. Think of it as the object's primary key.
  </div>
  <div class="vocab-row">
    <span class="vocab-term">Type</span><span class="vocab-dash">—</span>
    int, list, str, etc. Fixed for the object's life.
  </div>
  <div class="vocab-row">
    <span class="vocab-term">Value</span><span class="vocab-dash">—</span>
    the actual data inside.
  </div>
</div>

<div class="note-box">💡 A <strong>name</strong> (variable) is separate from all of this. Assignment (<code style="background:#313244;padding:2px 6px;border-radius:4px;color:#a6e3a1;font-family:'Courier New',monospace">x = obj</code>) doesn't copy the object — it <strong>binds</strong> the name <code style="background:#313244;padding:2px 6px;border-radius:4px;color:#a6e3a1;font-family:'Courier New',monospace">x</code> to the object. This is why people say Python has "names and bindings," not "variables and values."</div>

<h3 class="sub">🔹 Mutable vs. Immutable — The Core Split</h3>

<div class="mut-grid">
  <div class="mut-card mc-imm">
    <div class="mc-title">🔒 Immutable</div>
    <div class="mc-body">
      Cannot be changed after creation.<br>
      If you "change" one, Python creates<br>
      a new object and re-points your<br>
      name at it.<br><br>
      <strong style="color:#89b4fa">Types:</strong><br>
      int, float, bool, str,<br>
      tuple, frozenset, bytes
    </div>
  </div>
  <div class="mut-card mc-mut">
    <div class="mc-title">🔓 Mutable</div>
    <div class="mc-body">
      Can be changed in place —<br>
      same identity, different contents.<br><br>
      <strong style="color:#a6e3a1">Types:</strong><br>
      list, dict, set,<br>
      and most custom class instances
    </div>
  </div>
</div>

<div class="info-box">
  💡 <strong>Why the split exists:</strong> Immutability lets Python safely share and cache objects, makes objects hashable (so they can be dict keys), and makes code easier to reason about. Mutability lets you modify large structures without the cost of copying. It's a deliberate trade-off, not an accident.
</div>

<div class="note-box">
  💡 <strong>SQL analogy for the split:</strong> an immutable object is like a primary-key value in a row — you don't edit it, you insert a new row. A mutable object is like a regular column you UPDATE in place. Same row (identity), new content.
</div>

<!-- ══ SECTION 2: EXAMPLES ══ -->
<div class="ex-header"><h2>2. Example (runnable, step by step)</h2></div>

<div class="ex-block">
  <div class="ex-title"><span class="ex-badge">Example 1</span> A name is a pointer; rebinding one name doesn't touch another</div>
  <div class="code-block" style="margin:0 0 8px 0">
a = <span class="cn">5</span><br>
b = a        <span class="cc"># b now points at the SAME object as a (the int 5)</span><br>
a = <span class="cn">99</span>      <span class="cc"># a is RE-POINTED to a new object (99). b is untouched.</span><br>
<span class="cm">print</span>(<span class="cs">"a:"</span>, a, <span class="cs">"b:"</span>, b)
  </div>
  <div class="output-block">a: 99 b: 5</div>
  <div class="step-box">
    <strong>Step by step:</strong><br>
    <code>a = 5</code> binds <code>a</code> to the int object 5.<br>
    <code>b = a</code> binds <code>b</code> to that same object.<br>
    <code>a = 99</code> does not modify the object 5 — int is immutable. It rebinds <code>a</code> to a new object 99.<br>
    <code>b</code> still points at 5.<br><br>
    This always works "intuitively" for immutables, which is exactly why the next example surprises people.
  </div>
</div>

<div class="ex-block">
  <div class="ex-title"><span class="ex-badge">Example 2</span> Aliasing a mutable object: one object, two names</div>
  <div class="code-block" style="margin:0 0 8px 0">
list1 = [<span class="cn">1</span>, <span class="cn">2</span>, <span class="cn">3</span>]<br>
list2 = list1         <span class="cc"># NOT a copy — list2 points at the SAME list</span><br>
list2.<span class="cm">append</span>(<span class="cn">4</span>)       <span class="cc"># mutate the object in place</span><br>
<span class="cm">print</span>(<span class="cs">"list1:"</span>, list1)<br>
<span class="cm">print</span>(<span class="cs">"list2:"</span>, list2)<br>
<span class="cm">print</span>(<span class="cs">"same object:"</span>, list1 <span class="ck">is</span> list2)
  </div>
  <div class="output-block">
list1: [1, 2, 3, 4]<br>
list2: [1, 2, 3, 4]<br>
same object: True
  </div>
  <div class="step-box">
    <strong>This is the moment the model earns its keep.</strong><br>
    <code>list2 = list1</code> copied the <em>pointer</em>, not the list. <code>.append(4)</code> mutated the single shared object, so both names see the change. <code>is</code> confirms they're literally the same object.<br><br>
    This is the <strong>#1 source of "why did my original data change?!" bugs</strong> in data work.
  </div>
</div>

<div class="ex-block">
  <div class="ex-title"><span class="ex-badge">Example 3</span> Function arguments: mutation vs. rebinding</div>
  <div class="code-block" style="margin:0 0 8px 0">
<span class="ck">def</span> <span class="cm">add_item</span>(seq):<br>
&nbsp;&nbsp;&nbsp;&nbsp;seq.<span class="cm">append</span>(<span class="cs">"new"</span>)       <span class="cc"># MUTATES the passed-in object</span><br>
<br>
data = [<span class="cs">"a"</span>, <span class="cs">"b"</span>]<br>
<span class="cm">add_item</span>(data)<br>
<span class="cm">print</span>(<span class="cs">"after mutate:"</span>, data)<br>
<br>
<span class="ck">def</span> <span class="cm">rebind</span>(seq):<br>
&nbsp;&nbsp;&nbsp;&nbsp;seq = [<span class="cs">"totally"</span>, <span class="cs">"new"</span>]   <span class="cc"># REBINDS the local name only</span><br>
<br>
data2 = [<span class="cs">"x"</span>, <span class="cs">"y"</span>]<br>
<span class="cm">rebind</span>(data2)<br>
<span class="cm">print</span>(<span class="cs">"after rebind:"</span>, data2)
  </div>
  <div class="output-block">
after mutate: ['a', 'b', 'new']<br>
after rebind: ['x', 'y']
  </div>
  <div class="step-box">
    This kills the "pass by value vs. pass by reference" question — it's the wrong frame for Python.<br><br>
    <strong>The truth:</strong> Python passes the binding. Inside the function, the parameter <code>seq</code> initially points at the same object the caller passed. If you <strong>mutate</strong> that object (<code>.append</code>), the caller sees it. If you <strong>rebind</strong> the local name (<code>seq = [...]</code>), you've only changed where the local name points; the caller's binding is untouched.<br><br>
    The official term is <strong>"call by object reference" / "call by sharing."</strong>
  </div>
</div>

<div class="ex-block">
  <div class="ex-title"><span class="ex-badge">Example 4</span> Copy vs. deepcopy</div>
  <div class="code-block" style="margin:0 0 8px 0">
<span class="ck">import</span> copy<br>
original = [[<span class="cn">1</span>, <span class="cn">2</span>], [<span class="cn">3</span>, <span class="cn">4</span>]]       <span class="cc"># a list containing two inner lists</span><br>
shallow  = copy.<span class="cm">copy</span>(original)      <span class="cc"># new outer list, SAME inner lists</span><br>
deep     = copy.<span class="cm">deepcopy</span>(original)  <span class="cc"># new outer list, NEW inner lists</span><br>
<br>
original[<span class="cn">0</span>].<span class="cm">append</span>(<span class="cn">999</span>)            <span class="cc"># mutate an inner list</span><br>
<span class="cm">print</span>(<span class="cs">"original:"</span>, original)<br>
<span class="cm">print</span>(<span class="cs">"shallow: "</span>, shallow)<br>
<span class="cm">print</span>(<span class="cs">"deep:    "</span>, deep)
  </div>
  <div class="output-block">
original: [[1, 2, 999], [3, 4]]<br>
shallow:  [[1, 2, 999], [3, 4]]<br>
deep:     [[1, 2], [3, 4]]
  </div>
  <div class="step-box">
    A <strong>shallow copy</strong> duplicates the outer container but the inner objects are still shared — so mutating an inner list shows up in the shallow copy.<br><br>
    A <strong>deep copy</strong> recursively duplicates everything, so it's fully independent.<br><br>
    This matters the instant you work with nested structures (lists of lists, dicts of lists — i.e., basically all real data).
  </div>
</div>

<!-- ══ SECTION 3: EDGE CASES ══ -->
<div class="warn-header"><h2>3. Edge Cases</h2></div>
<p style="color:#cdd6f4;font-family:'Segoe UI',sans-serif;font-size:0.93em;line-height:1.7;margin:0 0 14px 0">Six cases, all outputs verified by running the code.</p>

<div class="edge-block">
  <div class="edge-title"><span class="edge-badge">Edge 1</span> The Mutable Default Argument (the most famous Python trap)</div>
  <div class="code-block" style="margin:0 0 8px 0">
<span class="ck">def</span> <span class="cm">add_to</span>(item, target=[]):<br>
&nbsp;&nbsp;&nbsp;&nbsp;target.<span class="cm">append</span>(item)<br>
&nbsp;&nbsp;&nbsp;&nbsp;<span class="ck">return</span> target<br>
<br>
<span class="cm">print</span>(<span class="cm">add_to</span>(<span class="cn">1</span>))<br>
<span class="cm">print</span>(<span class="cm">add_to</span>(<span class="cn">2</span>))<br>
<span class="cm">print</span>(<span class="cm">add_to</span>(<span class="cn">3</span>))
  </div>
  <div class="output-block">
[1]<br>
[1, 2]<br>
[1, 2, 3]
  </div>
  <div class="why-box">
    <strong>Why it matters:</strong> the default value <code>[]</code> is created <em>once</em>, when the function is defined — not each time it's called. So every call that relies on the default shares the same list object. It accumulates across calls. This is a real bug that ships to production. (The fix is the headline Golden Rule in Part 2.)
  </div>
</div>

<div class="edge-block">
  <div class="edge-title"><span class="edge-badge">Edge 2</span> Small-Integer Caching, and Why the Demo is Subtle</div>
  <div class="code-block" style="margin:0 0 8px 0">
a = <span class="cn">256</span><br>
b = <span class="cn">256</span><br>
<span class="cm">print</span>(<span class="cs">"256 is 256:"</span>, a <span class="ck">is</span> b)   <span class="cc"># True</span>
  </div>
  <div class="output-block">256 is 256: True</div>
  <div class="why-box">
    <strong>Why it matters:</strong> CPython pre-creates and caches every integer from -5 to 256 at startup. Any name pointing at, say, 256 points at the one cached object — so <code>is</code> returns True. Outside that range, integers are normally separate objects.
  </div>
  <div class="nuance-box">
    <strong>Honest nuance:</strong> if you write <code>c = 257; d = 257</code> in the same script or function body, you may still get <code>c is d → True</code>, because the compiler folds identical literals in one code block into one constant. To see the "real" False, you need them created separately:
  </div>
  <div class="code-block" style="margin:0 0 8px 0">
c = <span class="cn">257</span><br>
d = <span class="cm">int</span>(<span class="cs">"257"</span>)     <span class="cc"># forces a fresh object, defeats literal-folding</span><br>
<span class="cm">print</span>(<span class="cs">"257 is 257:"</span>, c <span class="ck">is</span> d)   <span class="cc"># False</span>
  </div>
  <div class="output-block">257 is 257: False</div>
  <div class="why-box">
    The takeaway isn't the cache boundary trivia — it's the rule it teaches: <strong>never use <code>is</code> to compare values. Use <code>==</code>.</strong> <code>is</code> is for identity only (and for None).
  </div>
</div>

<div class="edge-block">
  <div class="edge-title"><span class="edge-badge">Edge 3</span> String Interning</div>
  <div class="code-block" style="margin:0 0 8px 0">
s1 = <span class="cs">"hello"</span><br>
s2 = <span class="cs">"hello"</span><br>
<span class="cm">print</span>(s1 <span class="ck">is</span> s2)   <span class="cc"># True</span>
  </div>
  <div class="output-block">True</div>
  <div class="why-box">
    <strong>Why it matters:</strong> Python interns (caches) certain strings — short ones, and identifier-like literals — so identical literals share one object. But you can't rely on which strings get interned; it varies by content, Python version, and whether the strings are literals in the same compiled block vs. built at runtime (e.g., from user input or concatenation). Same lesson as Edge 2: <strong>comparing strings with <code>is</code> is a latent bug. Use <code>==</code>.</strong>
  </div>
</div>

<div class="edge-block">
  <div class="edge-title"><span class="edge-badge">Edge 4</span> A Tuple Containing a List (the "immutable isn't always frozen" case)</div>
  <div class="code-block" style="margin:0 0 8px 0">
t = (<span class="cn">1</span>, <span class="cn">2</span>, [<span class="cn">3</span>, <span class="cn">4</span>])<br>
t[<span class="cn">2</span>].<span class="cm">append</span>(<span class="cn">5</span>)        <span class="cc"># allowed — mutating the inner list</span><br>
<span class="cm">print</span>(<span class="cs">"after:"</span>, t)<br>
<br>
t[<span class="cn">0</span>] = <span class="cn">99</span>             <span class="cc"># not allowed — rebinding a tuple slot</span>
  </div>
  <div class="output-block">
after: (1, 2, [3, 4, 5])<br>
TypeError: 'tuple' object does not support item assignment
  </div>
  <div class="why-box">
    <strong>Why it matters:</strong> a tuple's immutability is <em>shallow</em>. It freezes <em>which objects</em> its slots point at — you can't swap slot [0] for 99. But if a slot points at a mutable object, that object can still be changed. "Immutable container" ≠ "everything inside is immutable." This also means such a tuple is <strong>not hashable</strong> and can't be a dict key.
  </div>
</div>

<div class="edge-block">
  <div class="edge-title"><span class="edge-badge">Edge 5</span> == vs is, Side by Side</div>
  <div class="code-block" style="margin:0 0 8px 0">
l1 = [<span class="cn">1</span>, <span class="cn">2</span>, <span class="cn">3</span>]<br>
l2 = [<span class="cn">1</span>, <span class="cn">2</span>, <span class="cn">3</span>]<br>
<span class="cm">print</span>(<span class="cs">"l1 == l2:"</span>, l1 == l2)   <span class="cc"># equal VALUES</span><br>
<span class="cm">print</span>(<span class="cs">"l1 is l2:"</span>, l1 <span class="ck">is</span> l2)   <span class="cc"># different OBJECTS</span>
  </div>
  <div class="output-block">
l1 == l2: True<br>
l1 is l2: False
  </div>
  <div class="why-box">
    <strong>Why it matters:</strong> the cleanest statement of the whole distinction. <code>==</code> asks "same value?" <code>is</code> asks "same object in memory?" Two separately built lists are equal but not identical. Mixing these up is one of the most common interview gotchas and real bugs.
  </div>
</div>

<div class="edge-block">
  <div class="edge-title"><span class="edge-badge">Edge 6</span> Augmented Assignment (+=) Behaves Differently on list vs. tuple</div>
  <div class="code-block" style="margin:0 0 8px 0">
lst = [<span class="cn">1</span>, <span class="cn">2</span>]<br>
<span class="cm">print</span>(<span class="cs">"list id before:"</span>, <span class="cm">id</span>(lst))<br>
lst += [<span class="cn">3</span>]                        <span class="cc"># mutates IN PLACE — same object</span><br>
<span class="cm">print</span>(<span class="cs">"list id after: "</span>, <span class="cm">id</span>(lst))<br>
<br>
tup = (<span class="cn">1</span>, <span class="cn">2</span>)<br>
<span class="cm">print</span>(<span class="cs">"tuple id before:"</span>, <span class="cm">id</span>(tup))<br>
tup += (<span class="cn">3</span>,)                        <span class="cc"># creates a NEW object — rebinds</span><br>
<span class="cm">print</span>(<span class="cs">"tuple id after: "</span>, <span class="cm">id</span>(tup))
  </div>
  <div class="output-block">
list id before: 139722336735104<br>
list id after:  139722336735104 &nbsp;&nbsp; &lt;- SAME id<br>
tuple id before: 139722334685504<br>
tuple id after:  139722334804416 &nbsp; &lt;- DIFFERENT id<br>
<span style="color:#6c7086">(ids will differ per run, but the pattern holds)</span>
  </div>
  <div class="why-box">
    <strong>Why it matters:</strong> for a mutable list, <code>+=</code> calls <code>__iadd__</code> and mutates in place (same identity) — so it behaves like <code>.extend()</code> and affects all aliases. For an immutable tuple, <code>+=</code> can't mutate, so it builds a new object and rebinds the name. <strong>Same operator, opposite consequences depending on mutability.</strong>
  </div>
</div>

<div class="edge-block">
  <div class="edge-title"><span class="edge-badge">Edge 6b</span> The Genuinely Cursed One: += Inside a Tuple</div>
  <div class="code-block" style="margin:0 0 8px 0">
t = (<span class="cn">1</span>, <span class="cn">2</span>, [<span class="cn">3</span>, <span class="cn">4</span>])<br>
t[<span class="cn">2</span>] += [<span class="cn">5</span>]
  </div>
  <div class="output-block">
TypeError: 'tuple' object does not support item assignment<br>
&gt;&gt;&gt; print(t)<br>
(1, 2, [3, 4, 5])
  </div>
  <div class="why-box">
    <strong>Why it matters:</strong> this <em>both succeeds and fails</em>. <code>t[2] += [5]</code> is <code>t[2] = t[2].__iadd__([5])</code>. The <code>__iadd__</code> part mutates the inner list (so 5 does get added), but then Python tries to reassign <code>t[2]</code> back into the tuple slot, which raises TypeError. End state: the list was modified and you got an exception.<br><br>
    It's the perfect demonstration that <code>+=</code> is <strong>mutate-then-rebind</strong>, and that tuple immutability only blocks the rebind. <strong>Senior-level interview material.</strong>
  </div>
</div>

<h2> <div class="info-box"><strong>Part 2:</strong> Golden Rules → Common Traps → Exercise</div> </h2>

<div class="part-header"><h2>4. Golden Rules</h2></div>

<p style="color:#cdd6f4;font-family:'Segoe UI',sans-serif;font-size:0.93em;line-height:1.7;margin:0 0 14px 0">Production patterns to memorize. Each one is a direct consequence of the object model from Part 1.</p>

<div class="rule-block">
  <div class="rule-title"><span class="rule-badge">Rule 1</span> Never use a mutable default argument. Use <code style="background:#313244;padding:2px 6px;border-radius:4px;color:#a6e3a1">None</code> as the sentinel.</div>
  <div class="code-block" style="margin:0 0 8px 0">
<span class="cc"># WRONG</span><br>
<span class="ck">def</span> <span class="cm">f</span>(target=[]):  ...<br>
<br>
<span class="cc"># RIGHT</span><br>
<span class="ck">def</span> <span class="cm">f</span>(target=<span class="ck">None</span>):<br>
&nbsp;&nbsp;&nbsp;&nbsp;<span class="ck">if</span> target <span class="ck">is</span> <span class="ck">None</span>:<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;target = []
  </div>
  <div class="verified-box">
    Verified — the fixed version resets per call:<br>
    add_to_fixed(1) -&gt; [1]<br>
    add_to_fixed(2) -&gt; [2] &nbsp;&nbsp; # no accumulation
  </div>
  <div class="rule-body" style="margin-top:8px">The default is evaluated once at definition time. <code>None</code> is immutable and safe; you build a fresh list inside the call.</div>
</div>

<div class="rule-block">
  <div class="rule-title"><span class="rule-badge">Rule 2</span> <code style="background:#313244;padding:2px 6px;border-radius:4px;color:#a6e3a1">is</code> is for identity and <code style="background:#313244;padding:2px 6px;border-radius:4px;color:#a6e3a1">None</code> only. <code style="background:#313244;padding:2px 6px;border-radius:4px;color:#a6e3a1">==</code> is for values.</div>
  <div class="rule-body">
    Use <code>if x is None:</code>, never <code>if x == None:</code>.<br>
    Never use <code>is</code> to compare numbers or strings — Part 1's caching edge cases are exactly why.<br>
    Python will even warn you (<code>SyntaxWarning: "is" with 'int' literal</code>).
  </div>
</div>

<div class="rule-block">
  <div class="rule-title"><span class="rule-badge">Rule 3</span> Copying a list: <code style="background:#313244;padding:2px 6px;border-radius:4px;color:#a6e3a1">b = a</code> is not a copy.</div>
  <div class="rule-body">
    Use <code>a[:]</code>, <code>a.copy()</code>, or <code>list(a)</code>.<br>
    And the instant you have nesting, those three are still <strong>shallow</strong> — reach for <code>copy.deepcopy()</code>.
  </div>
</div>

<div class="rule-block">
  <div class="rule-title"><span class="rule-badge">Rule 4</span> Mutable class attributes are shared across all instances.</div>
  <div class="rule-body" style="margin-bottom:8px">Put mutable state in <code>__init__</code> (instance attribute), never as a bare class attribute.</div>
  <div class="code-block" style="margin:0">
<span class="cc"># WRONG: items = [] at class level (shared)</span><br>
<span class="cc"># RIGHT:</span><br>
<span class="ck">class</span> <span class="cm">Cart</span>:<br>
&nbsp;&nbsp;&nbsp;&nbsp;<span class="ck">def</span> <span class="cm">__init__</span>(self):<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;self.items = []   <span class="cc"># fresh per instance</span>
  </div>
</div>

<div class="rule-block">
  <div class="rule-title"><span class="rule-badge">Rule 5</span> If a function mutates its argument, that's a side effect — make it obvious or avoid it.</div>
  <div class="rule-body">
    Either name it clearly (<code>append_processed_row</code>) and document it, or return a new object and don't touch the input.<br>
    Silent mutation of caller data is the bug that wastes the most debugging hours in data pipelines.
  </div>
</div>

<div class="rule-block">
  <div class="rule-title"><span class="rule-badge">Rule 6</span> Want a hashable / dict-key / set-member object? It must be immutable all the way down.</div>
  <div class="rule-body">
    A tuple of ints works; a tuple containing a list does not (Part 1, Edge 4).
  </div>
</div>

<div class="trap-header"><h2>5. Common Traps</h2></div>

<p style="color:#cdd6f4;font-family:'Segoe UI',sans-serif;font-size:0.93em;line-height:1.7;margin:0 0 14px 0">The bugs that bite — beginners and seniors. All outputs verified.</p>

<div class="trap-block">
  <div class="trap-title"><span class="trap-badge">Trap 1</span> Silent argument mutation corrupts your source data.</div>
  <div class="code-block" style="margin:0 0 8px 0">
<span class="ck">def</span> <span class="cm">normalize</span>(row):<br>
&nbsp;&nbsp;&nbsp;&nbsp;row.<span class="cm">append</span>(<span class="cs">"processed"</span>)<br>
&nbsp;&nbsp;&nbsp;&nbsp;<span class="ck">return</span> row<br>
<br>
master = [<span class="cs">"id1"</span>, <span class="cs">"value"</span>]<br>
result = <span class="cm">normalize</span>(master)<br>
<span class="cm">print</span>(master)   <span class="cc"># ['id1', 'value', 'processed']  ← master is wrecked</span>
  </div>
  <div class="output-block">['id1', 'value', 'processed']</div>
  <div class="why-box">
    You thought <code>result</code> was a new thing. It's the same object as <code>master</code>. This is the pandas-relevant one: it's the pure-Python version of the <code>SettingWithCopyWarning</code> / "did I modify a view or a copy?" confusion you hit in DataFrames. Same root cause — aliasing a mutable object.
  </div>
</div>

<div class="trap-block">
  <div class="trap-title"><span class="trap-badge">Trap 2</span> Aliasing vs. copying.</div>
  <div class="code-block" style="margin:0 0 8px 0">
b = a        <span class="cc"># alias: b tracks every change to a</span><br>
c = a[:]     <span class="cc"># copy: independent</span>
  </div>
  <div class="why-box">
    After <code>a.append(4)</code>: <code>b</code> → <code>[1,2,3,4]</code>, <code>c</code> → <code>[1,2,3]</code>.<br>
    If you didn't mean to share, you created a bug.
  </div>
</div>

<div class="trap-block">
  <div class="trap-title"><span class="trap-badge">Trap 3</span> Mutable class attribute shared across instances.</div>
  <div class="code-block" style="margin:0 0 8px 0">
<span class="ck">class</span> <span class="cm">Cart</span>:<br>
&nbsp;&nbsp;&nbsp;&nbsp;items = []           <span class="cc"># SHARED by every Cart</span><br>
&nbsp;&nbsp;&nbsp;&nbsp;<span class="ck">def</span> <span class="cm">add</span>(self, x): self.items.<span class="cm">append</span>(x)<br>
<br>
c1, c2 = <span class="cm">Cart</span>(), <span class="cm">Cart</span>()<br>
c1.<span class="cm">add</span>(<span class="cs">"apple"</span>)<br>
<span class="cm">print</span>(c2.items)   <span class="cc"># ['apple']  ← c2 has c1's data</span>
  </div>
  <div class="output-block">['apple']</div>
  <div class="why-box">
    <code>c2</code> never added anything, yet it sees "apple". One list object backs all instances. Verified.
  </div>
</div>

<div class="trap-block">
  <div class="trap-title"><span class="trap-badge">Trap 4</span> The <code style="background:#3e1818;padding:2px 6px;border-radius:4px;color:#f38ba8">[[0]*3]*3</code> grid trap.</div>
  <div class="code-block" style="margin:0 0 8px 0">
grid = [[<span class="cn">0</span>]*<span class="cn">3</span> <span class="ck">for</span> _ <span class="ck">in</span> <span class="cm">range</span>(<span class="cn">3</span>)]   <span class="cc"># GOOD: 3 distinct rows</span><br>
bad  = [[<span class="cn">0</span>]*<span class="cn">3</span>]*<span class="cn">3</span>                    <span class="cc"># BAD: 3 aliases of ONE row</span><br>
<br>
grid[<span class="cn">0</span>][<span class="cn">0</span>] = <span class="cn">1</span>   <span class="cc"># -> [[1,0,0],[0,0,0],[0,0,0]]</span><br>
bad[<span class="cn">0</span>][<span class="cn">0</span>]  = <span class="cn">1</span>   <span class="cc"># -> [[1,0,0],[1,0,0],[1,0,0]]  all rows change</span>
  </div>
  <div class="output-block">
grid: [[1,0,0],[0,0,0],[0,0,0]]<br>
bad:  [[1,0,0],[1,0,0],[1,0,0]]
  </div>
  <div class="why-box">
    <code>* 3</code> on a list of lists copies the pointer three times. Setting one cell appears to set a whole column. Extremely common when building matrices/boards by hand — relevant to any from-scratch ML or DP grid work.
  </div>
</div>

<div class="trap-block">
  <div class="trap-title"><span class="trap-badge">Trap 5</span> Late binding closures in a loop.</div>
  <div class="code-block" style="margin:0 0 8px 0">
funcs = [<span class="ck">lambda</span>: i <span class="ck">for</span> i <span class="ck">in</span> <span class="cm">range</span>(<span class="cn">3</span>)]<br>
<span class="cm">print</span>([f() <span class="ck">for</span> f <span class="ck">in</span> funcs])   <span class="cc"># [2, 2, 2]  not [0, 1, 2]</span>
  </div>
  <div class="output-block">[2, 2, 2]</div>
  <div class="why-box">
    The lambdas don't capture the <em>value</em> of <code>i</code> — they capture the <em>name</em> <code>i</code>, which all three share. By the time you call them, the loop has finished and <code>i</code> is <code>2</code>.<br><br>
    Fix: <code>lambda i=i: i</code> to bind at definition time.<br><br>
    This is a Session 5 closure topic, but it's rooted in the same names-vs-objects idea, so meet it now.
  </div>
</div>

<div style="background:#1e1e2e; border-left:4px solid #cba6f7; padding:16px 20px; border-radius:8px; font-family:monospace; color:#cdd6f4">
  <h4 style="color:#cba6f7; margin:0 0 12px 0;">🔑 Key Takeaways</h4>
  <ul style="margin:0; padding-left:20px; line-height:2.2">
    <li>A name is a tag on an object — <code>=</code> binds a tag, it never copies the object</li>
    <li>Mutating a shared object is visible through every tag on it</li>
    <li>Immutable objects can't be mutated, so "changing" them rebinds the tag to a new object instead</li>
    <li><code>is</code> = same object, <code>==</code> = same value</li>
  </ul>
</div>

<div class="ml-header"><h2>7. ML Real-World Connection</h2></div>

<p style="color:#cdd6f4;font-family:'Segoe UI',sans-serif;font-size:0.93em;line-height:1.7;margin:0 0 14px 0">This isn't abstract trivia — mutability bugs are some of the nastiest in ML because they're <strong style="color:#f38ba8">silent</strong>. No error, just wrong numbers. Three places it bites in real pipelines:</p>

<div class="ml-block">
  <div class="ml-block-title"><span class="ml-badge">ML Bite 1</span> Accidental data corruption / label leakage via aliasing</div>
  <p style="color:#cdd6f4;font-family:'Segoe UI',sans-serif;font-size:0.92em;line-height:1.7;margin:0 0 8px 0">You write a feature-engineering function that "returns" a transformed row but actually mutates the input:</p>
  <div class="code-block" style="margin:0 0 8px 0">
<span class="ck">def</span> <span class="cm">add_features</span>(row):<br>
&nbsp;&nbsp;&nbsp;&nbsp;row.<span class="cm">append</span>(row[<span class="cn">0</span>] * <span class="cn">2</span>)   <span class="cc"># mutates the ORIGINAL</span><br>
&nbsp;&nbsp;&nbsp;&nbsp;<span class="ck">return</span> row<br>
<br>
raw = [<span class="cn">3</span>, <span class="cn">5</span>]                  <span class="cc"># [feature, label]</span><br>
features = <span class="cm">add_features</span>(raw)<br>
<span class="cc"># raw is now [3, 5, 6] -> your "raw" dataset is silently changed</span>
  </div>
  <div class="output-block">raw corrupted: [3, 5, 6]</div>
  <div class="verified-box">
    ✅ Verified output: raw corrupted: [3, 5, 6]<br>
    In a real pipeline this is how you accidentally leak or destroy data between train/test splits.<br>
    The fix is Golden Rule 5: make transforms pure (<code>return row + [row[0]*2]</code>).
  </div>
</div>

<div class="ml-block">
  <div class="ml-block-title"><span class="ml-badge">ML Bite 2</span> Shared config / hyperparameter dicts persisting across runs</div>
  <p style="color:#cdd6f4;font-family:'Segoe UI',sans-serif;font-size:0.92em;line-height:1.7;margin:0 0 8px 0">The mutable-default trap wearing a data-science costume:</p>
  <div class="code-block" style="margin:0 0 8px 0">
<span class="ck">def</span> <span class="cm">train</span>(params={<span class="cs">"lr"</span>: <span class="cn">0.01</span>}):<br>
&nbsp;&nbsp;&nbsp;&nbsp;params[<span class="cs">"lr"</span>] *= <span class="cn">0.1</span><br>
&nbsp;&nbsp;&nbsp;&nbsp;<span class="ck">return</span> params<br>
<br>
<span class="cm">train</span>()   <span class="cc"># {'lr': 0.001}</span><br>
<span class="cm">train</span>()   <span class="cc"># {'lr': 0.0001}  ← decayed AGAIN across calls</span>
  </div>
  <div class="output-block">
{'lr': 0.001}<br>
{'lr': 0.0001}
  </div>
  <div class="verified-box">
    ✅ Verified. Your learning rate silently compounds between experiments and your results aren't reproducible.<br>
    This is exactly the kind of bug that makes "I ran the same code and got different numbers" tickets.<br>
    Fix: <code>params=None</code> sentinel.
  </div>
</div>

<div class="ml-block">
  <div class="ml-block-title"><span class="ml-badge">ML Bite 3</span> NumPy/pandas views vs. copies — the SettingWithCopyWarning root cause</div>
  <p style="color:#cdd6f4;font-family:'Segoe UI',sans-serif;font-size:0.92em;line-height:1.7;margin:0 0 8px 0">This is the same concept behind NumPy/pandas views vs. copies — the thing you already hit as <code style="background:#313244;padding:2px 6px;border-radius:4px;color:#a6e3a1;font-family:'Courier New',monospace">SettingWithCopyWarning</code>. NumPy slicing returns a view that shares memory with the original:</p>
  <div class="code-block" style="margin:0 0 8px 0">
a = np.<span class="cm">array</span>([<span class="cn">1</span>, <span class="cn">2</span>, <span class="cn">3</span>, <span class="cn">4</span>])<br>
view = a[<span class="cn">1</span>:<span class="cn">3</span>]    <span class="cc"># shares memory</span><br>
view[<span class="cn">0</span>] = <span class="cn">99</span>     <span class="cc"># a becomes [1, 99, 3, 4]  ← original changed</span>
  </div>
  <div class="output-block">a: [1, 99, 3, 4]</div>
  <div class="verified-box">✅ Verified. This is pure-Python aliasing at the array level.</div>
  <div class="why-box">
    When pandas warns you "you're setting on a copy of a slice," it's warning about exactly the names-vs-objects ambiguity from Part 1. You now understand the root cause of a warning you've been working around in your Commercial Wire data work — it's not a pandas quirk, it's the Python object model. That connection alone earns this session.
  </div>
</div>

<div class="interview-header"><h2>8. Interview Questions</h2></div>

<div class="sub-header"><h3>8a — Conceptual Q&amp;A (12 questions)</h3></div>

<div class="qa-block">
  <div class="qa-q"><span class="q-num">1</span> "Is Python pass-by-value or pass-by-reference?"</div>
  <div class="qa-a">
    <span class="ans-label">A</span>
    Neither. It's <strong>call-by-object-reference</strong> (call by sharing). The function receives a binding to the same object. Mutating the object is visible to the caller; rebinding the local name is not.
  </div>
</div>

<div class="qa-block">
  <div class="qa-q"><span class="q-num">2</span> "What's the difference between <code>is</code> and <code>==</code>?"</div>
  <div class="qa-a">
    <span class="ans-label">A</span>
    <code>is</code> compares <strong>identity</strong> (same object in memory, i.e. same <code>id()</code>).<br>
    <code>==</code> compares <strong>value</strong> (via <code>__eq__</code>).<br>
    Two distinct lists with equal contents are <code>== True</code> but <code>is False</code>.
  </div>
</div>

<div class="qa-block">
  <div class="qa-q"><span class="q-num">3</span> "Name the immutable built-in types."</div>
  <div class="qa-a">
    <span class="ans-label">A</span>
    <strong>Immutable:</strong> <code>int</code>, <code>float</code>, <code>bool</code>, <code>complex</code>, <code>str</code>, <code>tuple</code>, <code>frozenset</code>, <code>bytes</code>, and <code>None</code>.<br>
    <strong>Mutable:</strong> <code>list</code>, <code>dict</code>, <code>set</code>, <code>bytearray</code>, most custom objects.
  </div>
</div>

<div class="qa-block">
  <div class="qa-q"><span class="q-num">4</span> "Why is <code>def f(x=[])</code> dangerous?"</div>
  <div class="qa-a">
    <span class="ans-label">A</span>
    The default is created <strong>once at definition time</strong> and shared across all calls relying on it, so it accumulates state. Use <code>x=None</code> and build the list inside.
  </div>
</div>

<div class="qa-block">
  <div class="qa-q"><span class="q-num">5</span> "Can a tuple ever 'change'?"</div>
  <div class="qa-a">
    <span class="ans-label">A</span>
    Its bindings can't — you can't reassign a slot. But if a slot holds a mutable object (like a list), that object can be mutated. So a tuple's immutability is <strong>shallow</strong>.
  </div>
</div>

<div class="qa-block">
  <div class="qa-q"><span class="q-num">6</span> "Why must dict keys / set members be immutable?"</div>
  <div class="qa-a">
    <span class="ans-label">A</span>
    They must be <strong>hashable</strong>, and hashability requires a stable hash over the object's lifetime. Mutable objects could change and break the hash table's invariants, so <code>list</code> and <code>dict</code> are unhashable.
  </div>
</div>

<div class="qa-block">
  <div class="qa-q"><span class="q-num">7</span> "Difference between shallow and deep copy?"</div>
  <div class="qa-a">
    <span class="ans-label">A</span>
    <strong>Shallow</strong> (<code>copy.copy</code>, <code>a[:]</code>, <code>a.copy()</code>, <code>list(a)</code>) duplicates the outer container but shares the inner objects.<br>
    <strong>Deep</strong> (<code>copy.deepcopy</code>) recursively duplicates everything.<br>
    Matters the moment data is nested.
  </div>
</div>

<div class="qa-block">
  <div class="qa-q"><span class="q-num">8</span> "Why does <code>a is b</code> sometimes return True for two separate int or str assignments?"</div>
  <div class="qa-a">
    <span class="ans-label">A</span>
    CPython caches small ints (-5 to 256) and interns some strings, so identical literals may share one object. It's an implementation detail — never rely on it; use <code>==</code>.
  </div>
</div>

<div class="qa-block">
  <div class="qa-q"><span class="q-num">9</span> "What does <code>id()</code> return and when does it change?"</div>
  <div class="qa-a">
    <span class="ans-label">A</span>
    A unique identifier for the object (in CPython, its memory address), constant for the object's lifetime. A new object gets a new id. Rebinding a name doesn't change any object's id.
  </div>
</div>

<div class="qa-block">
  <div class="qa-q"><span class="q-num">10</span> "After <code>b = a</code> where a is a list, how do you check if they're the same object vs. equal?"</div>
  <div class="qa-a">
    <span class="ans-label">A</span>
    <code>a is b</code> → same object (True here).<br>
    <code>a == b</code> → equal value.<br>
    To get an independent copy, use <code>b = a.copy()</code>.
  </div>
</div>

<div class="qa-block">
  <div class="qa-q"><span class="q-num">11</span> "What happens with <code>t[2] += [5]</code> when <code>t = (1,2,[3,4])</code>?"</div>
  <div class="qa-a">
    <span class="ans-label">A</span>
    It <strong>both succeeds and raises</strong>. <code>+=</code> mutates the inner list in place (so 5 is added), then tries to reassign the tuple slot, which raises <code>TypeError</code>. End state: list modified and exception raised. Demonstrates <code>+=</code> is <strong>mutate-then-rebind</strong>.
  </div>
</div>

<div class="qa-block">
  <div class="qa-q"><span class="q-num">12</span> "How does this concept relate to pandas <code>SettingWithCopyWarning</code>?"</div>
  <div class="qa-a">
    <span class="ans-label">A</span>
    Same root: ambiguity over whether you hold a <strong>view</strong> (alias, shared memory) or a <strong>copy</strong> (independent). The warning flags that your assignment might be hitting a copy, so it may not affect the original — the names-vs-objects problem at the DataFrame level.
  </div>
</div>
